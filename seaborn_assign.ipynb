# Que 3:
Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head(3)
```

```python
# Box plot for 'age' by 'pclass'
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
sns.boxplot(data=titanic, x="pclass", y="age", ax=axes[0])
axes[0].set_title("Box Plot of Age by Pclass")
axes[0].set_xlabel("Pclass")
axes[0].set_ylabel("Age")

# Box plot for 'fare' by 'pclass'
sns.boxplot(data=titanic, x="pclass", y="fare", ax=axes[1])
axes[1].set_title("Box Plot of Fare by Pclass")
axes[1].set_xlabel("Pclass")
axes[1].set_ylabel("Fare")
plt.tight_layout()
plt.show()
```

# Que 5:
Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris_data = sns.load_dataset("iris")

# Create a pair plot with hue for the 'species' column
sns.set(style="ticks") 
sns.pairplot(iris_data, hue="species")

plt.suptitle("Pair Plot of Iris Dataset with Hue for Species", y=1.02)
plt.show()
```

# Que 6:
Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights_data = sns.load_dataset("flights")

flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

plt.figure(figsize=(10, 6)) 

sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d")

plt.title("Heatmap of Flights Dataset")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
