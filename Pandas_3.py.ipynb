# Q6.
### What do you understand about the windows function in pandas and list the types of windows functions?

# Ans.

In pandas, a window function, also known as a rolling or moving function, allows you to perform calculations on a set of data points defined by a window or a rolling window. These functions are especially useful for time series data or any data that has a natural ordering.

Types of window functions in pandas:

Rolling Functions: Rolling functions allow you to compute statistics (e.g., mean, sum, standard deviation) over a rolling window of a specified size. You can create a rolling window object using the rolling() method on a pandas Series or DataFrame and then apply various aggregation functions to it.

Expanding Functions: Expanding functions calculate statistics that gradually include more data points as the window expands over time. This means the size of the window increases with each data point, considering all data from the beginning up to the current point.

Exponentially Weighted Functions: Exponentially weighted functions give more weight to recent data points and less weight to older data points. They are commonly used for exponential moving averages, where more recent data points contribute more to the average.

Time-based Functions: Time-based functions allow you to work with time series data and handle datetime indices. They enable you to perform various operations like resampling, time shifting, and time aggregation over specified windows.
