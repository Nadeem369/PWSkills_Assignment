# Q1:
What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

### Line Plot

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x,y, marker='o')
```

### Scatter Plot

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
```

```python
plt.scatter(x,y)
```

### Bar Plot

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.bar(x,y)
```

### Histogram

```python
data = [3, 5, 5, 7, 8, 8, 8, 9, 9, 9, 10, 10, 10, 11]
plt.hist(data, bins=5, edgecolor='b')
```

### Pie Chart

```python
labels = ['Apples', 'Bananas', 'Grapes', 'Oranges']
sizes = [30, 25, 20, 25]
```

```python
plt.pie(sizes, labels=labels)
```

# Q2:
What is a scatter plot? Use the following code to generate data for x and y.

```python
import numpy as np
```

```python
np.random.seed(3)
```

```python
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel("This is my X")
plt.ylabel("This is my Y")
plt.title("This is my Title")
plt.show()
```

# Q3.
Why is the subplot() function used? Draw four line plots using the subplot() function.

```python
x = np.array([0, 1, 2, 3, 4, 5])

# Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])

# Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])

# Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])

# Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, subplot 1
plt.plot(x, y1)
plt.title('Line 1')
```

```python
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, subplot 2
plt.plot(x, y2)
plt.title('Line 2')
```

```python
plt.subplot(2, 2, 3)  # 2 rows, 2 columns, subplot 3
plt.plot(x, y3)
plt.title('Line 3')
```

```python
plt.subplot(2, 2, 4)  # 2 rows, 2 columns, subplot 4
plt.plot(x, y4)
plt.title('Line 4')
```

# Q5.
What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
data = [box1, box2]
```

```python
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.title('Box Plot of Box 1 and Box 2')
plt.ylabel('Value')
plt.show()
```
